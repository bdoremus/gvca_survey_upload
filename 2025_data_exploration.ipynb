```python
import pandas as pd

df = pd.read_csv("/workspaces/gvca_survey_analytics/2024 Parent Satisfaction Survey.csv")
pd.set_option("display.max_columns", None)
display(df.head())
```

```python
# Extract the first row
first_row = df.iloc[0]

# Create new headers by combining current headers with the first row values
new_headers = [(header, first_row[i]) for i, header in enumerate(df.columns)]

# Update the dataframe with new headers
df.columns = new_headers

# Drop the first row as it is now part of the headers
df = df.drop(0).reset_index(drop=True)

# Display the updated dataframe
display(df.head())
```

```python
# Update headers where the first value is 'Unnamed'
for i in range(1, len(new_headers)):
    if 'Unnamed' in new_headers[i][0]:
        new_headers[i] = (new_headers[i-1][0], new_headers[i][1])

# Update the dataframe with the new headers
df.columns = new_headers

# Display the updated dataframe
display(df.head())
```
